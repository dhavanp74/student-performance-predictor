## Feature Engineering 

```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)
```

```python
num_students = 2000
```

```python
study_hours = np.random.randint(1, 10, num_students)
```

```python
sleep_hours = np.random.randint(4, 10, num_students)
```

```python
attendance_percentage = np.random.randint(
    50,
    100,
    num_students
)
```

```python
assignments_completed = np.random.randint(
    0,
    11,
    num_students
)
```

```python
screen_time = np.random.randint(
    1,
    12,
    num_students
)
```

```python
coding_interest = np.random.randint(
    1,
    11,
    num_students
)
```

```python
logical_reasoning = np.random.randint(
    1,
    11,
    num_students
)
```

```python
creativity_score = np.random.randint(
    1,
    11,
    num_students
)
```

```python
communication_skill = np.random.randint(
    1,
    11,
    num_students
)
```

```python
stress_level = np.random.randint(
    1,
    11,
    num_students
)
```

```python
df = pd.DataFrame({
    "study_hours": study_hours,
    "sleep_hours": sleep_hours,
    "attendance_percentage": attendance_percentage,
    "assignments_completed": assignments_completed,
    "screen_time": screen_time,
    "coding_interest": coding_interest,
    "logical_reasoning": logical_reasoning,
    "creativity_score": creativity_score,
    "communication_skill": communication_skill,
    "stress_level": stress_level
})
```

```python
df.head()
```

```python
df.shape
```

```python
noise = np.random.normal(
    0,
    5,
    num_students
)
```

```python
performance_score = (
    study_hours * 5
    + attendance_percentage * 0.3
    + assignments_completed * 2
    + sleep_hours * 1.5
    - screen_time * 1.5
    - stress_level * 2
    + logical_reasoning * 2
    + noise
)
```

```python
performance_score = (
    (performance_score - performance_score.min())
    /
    (
        performance_score.max()
        - performance_score.min()
    )
) * 100
```

```python
df["performance_score"] = performance_score
```

```python
df.head()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))

sns.histplot(
    df["performance_score"],
    kde=True
)

plt.title("Performance Score Distribution")

plt.show()
```

```python
career_labels = []
```

```python
career_labels = []

for i in range(num_students):

    if (
        coding_interest[i] >= 8
        and logical_reasoning[i] >= 8
    ):
        career = "Data Science"

    elif (
        creativity_score[i] >= 8
        and communication_skill[i] >= 7
    ):
        career = "UI/UX Design"

    elif (
        logical_reasoning[i] >= 8
        and stress_level[i] <= 5
    ):
        career = "Cybersecurity"

    elif (
        coding_interest[i] >= 7
        and creativity_score[i] >= 7
    ):
        career = "Game Development"

    elif (
        coding_interest[i] >= 7
        and screen_time[i] >= 7
    ):
        career = "Mobile Development"

    elif (
        logical_reasoning[i] >= 7
        and attendance_percentage[i] >= 85
    ):
        career = "Cloud Computing"

    elif (
        coding_interest[i] >= 6
        and logical_reasoning[i] >= 6
    ):
        career = "Backend Development"

    elif (
        communication_skill[i] >= 7
        and logical_reasoning[i] >= 6
    ):
        career = "Software Testing"

    elif (
        coding_interest[i] >= 6
        and creativity_score[i] >= 6
    ):
        career = "Web Development"

    else:
        career = "DevOps Engineering"

    career_labels.append(career)
```

```python
df["career"] = career_labels
```

```python
df.head(20)
```

```python
print(df["career"].value_counts())
```

```python
plt.figure(figsize=(10,5))

sns.countplot(
    x="career",
    data=df
)

plt.xticks(rotation=15)

plt.title("Career Distribution")

plt.show()
```

#### Affinity Scores


```python
df["Data Science"] = (
    coding_interest * 0.35
    + logical_reasoning * 0.4
    + performance_score * 0.15
    + study_hours * 0.1
)
```

```python
df["Backend Development"] = (
    coding_interest * 0.4
    + logical_reasoning * 0.35
    + assignments_completed * 0.15
    + attendance_percentage * 0.1
)
```

```python
df["Cybersecurity"] = (
    logical_reasoning * 0.4
    + stress_level * (-0.2)
    + attendance_percentage * 0.2
    + performance_score * 0.2
)
```

```python
df["UI/UX Design"] = (
    creativity_score * 0.45
    + communication_skill * 0.3
    + screen_time * 0.15
    + coding_interest * 0.1
)
```

```python
df["Web Development"] = (
    coding_interest * 0.35
    + creativity_score * 0.3
    + communication_skill * 0.15
    + screen_time * 0.2
)
```

```python
df["Cloud Computing"] = (
    logical_reasoning * 0.35
    + attendance_percentage * 0.3
    + performance_score * 0.2
    + stress_level * (-0.15)
)
```

```python
df["Mobile Development"] = (
    coding_interest * 0.35
    + creativity_score * 0.25
    + screen_time * 0.25
    + communication_skill * 0.15
)
```

```python
df["Game Development"] = (
    creativity_score * 0.4
    + coding_interest * 0.3
    + screen_time * 0.2
    + logical_reasoning * 0.1
)
```

```python
df["DevOps Engineering"] = (
    logical_reasoning * 0.35
    + attendance_percentage * 0.25
    + stress_level * (-0.15)
    + assignments_completed * 0.25
)
```

```python
df["Software Testing"] = (
    logical_reasoning * 0.35
    + communication_skill * 0.25
    + assignments_completed * 0.2
    + stress_level * (-0.1)
    + coding_interest * 0.1
)
```

```python
career_columns = [
    "Data Science",
    "Backend Development",
    "Cybersecurity",
    "UI/UX Design",
    "Web Development",
    "Cloud Computing",
    "Mobile Development",
    "Game Development",
    "DevOps Engineering",
    "Software Testing"
]
```

```python
for col in career_columns:

    df[col] = (
        (df[col] - df[col].min())
        /
        (
            df[col].max()
            - df[col].min()
        )
    ) * 100
```

```python
df["recommended_career"] = (
    df[career_columns]
    .idxmax(axis=1)
)
```

```python
df[
    [
        "recommended_career"
    ] + career_columns
].head(20)
```

```python
print(df["recommended_career"].value_counts())
```

```python
plt.figure(figsize=(12,5))

sns.countplot(
    x="recommended_career",
    data=df
)

plt.xticks(rotation=20)

plt.title(
    "Recommended Career Distribution"
)

plt.show()
```
