#### Exploratotry Data Analysis (EDA)


```python
df["average_score"] = (
    df["math score"] +
    df["reading score"] +
    df["writing score"]
) / 3
```

```python
df.head(5)
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["average_score"], kde=True)

plt.title("Distribution of Average Scores")
plt.xlabel("Average Score")
plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize=(6,5))

sns.boxplot(
    x="gender",
    y="average_score",
    data=df
)

plt.title("Gender vs Average Score")

plt.show()

print("Female Count:", (df['gender'] == 'female').count())
print("Male Count:", (df['gender'] == 'female').count())
```

```python
plt.figure(figsize=(8,5))

sns.heatmap(
    df[[
        "math score",
        "reading score",
        "writing score",
        "average_score"
    ]].corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")

plt.show()
```

```python
plt.figure(figsize=(6,5))

sns.barplot(
    x="test preparation course",
    y="average_score",
    data=df
)

plt.title("Effect of Test Preparation on Scores")

plt.show()
```
